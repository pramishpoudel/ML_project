```python
df= pd.read_csv("creditcard.csv")
```

```python
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

Build a sample dataset containing similar distribution of normal transation and Fraudlent transaction

```python
legit_sample = legit.sample(n=492)
 
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns="Class",axis=1)
Y= new_dataset['Class']
```

```python
print(Y)
```

Train test split method  f

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
X_train.shape
```

Model Training

```python
model = LogisticRegression()
```

```python
 
model.fit(X_train,Y_train)
```

Model Evaluation with accuracy score

```python
X_train_prediction = model.predict(X_train)
tranning_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on  tranning data: ",tranning_data_accuracy)
```

```python
#accuracy on test data

X_test_prediction = model.predict(X_test)
test_data_prediction =accuracy_score(X_test_prediction,Y_test)
```
