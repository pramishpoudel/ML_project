Stemming the word 

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) #convert to the stemm words/root words
                       for word in stemmed_content 
                       if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
# fd['title']=fd['title'].apply(stemming)
fd['content']=fd['content'].apply(stemming)


 
```

```python
#convering the textual data to numerical data

vectorizer = TfidfVectorizer()
vectorizer.fit(X)
X=vectorizer.transform(X)
```

```python
#splitting the dataset to traning and test data

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)

# stratify=Y
# Ensures the distribution of classes in Y is preserved in both training and testing sets.
# Important for imbalanced datasets, e.g., if 90% are class 0 and 10% are class 1, stratifying keeps the same ratio in both sets.
# 5. random_state=2 A seed for the random number generator.Ensures that every time you run this code, you get the same split.
# Change the number (or remove it) → different random splits each time.
```

```python
# training logisticregression model
reg = LogisticRegression()
```

```python
reg.fit(X_train,Y_train)
```

```python
#evaluting the model accuracy
X_train_prediction =reg.predict(X_train)
tranning_accuracy_score = accuracy_score(X_train_prediction,Y_train)
```
